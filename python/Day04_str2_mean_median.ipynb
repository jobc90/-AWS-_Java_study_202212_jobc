```python
import math

def mean(num_list):
    #return sum(num_list)/len(num_list)
    for i in num_list:
        avr = sum(num_list)/len(num_list)
    return avr

def mean_2(num_list):
    sum_value = 0
    for i in num_list:
        sum_value = sum_value + i
    return sum_value/len(num_list)

game1 = csv_data[0]
print(mean(game1))

```

```python
# /는 나누기, //는 몫만 가져오고 나머지는 버린다.
def median(num_list):
    n = len(num_list)
    if n % 2 == 0:
        return (num_list[n//2 - 1] + num_list[n//2]) / 2
    else:
        return num_list[n//2] 

game1 = csv_data[0]
print(game1)
print(mean(game1))
print(mean_2(game1))
print(game1.sort())
print(game1)
print(game1[3])
print(median(game1))
print(median([5, 17, 26, 27, 35, 38]))

# assert함수는 해당 결과가 맞는지 표시해 주는 테스트 함수이다.
assert median([1, 5, 17, 26, 27, 35, 38]) == 26
```

```python
#편차
def dev(num_list):
    m = mean(num_list)
    return [x-m for x in num_list]
#분산
def var(num_list):
    n = len(num_list)
    d = dev(num_list)
    return sum([x*x for x in d])/(n-1)

#표준편차
def stdev(num_list):
    return math.sqrt(var(num_list))

#공분산
def covar(list_1, list_2):
    n = len(list_1)
    list_1_dev = dev(list_1)
    list_2_dev = dev(list_2)
    return sum(x * y for x, y in zip(list_1_dev, list_2_dev))/(n-1)

#상관도
def corel(list_1, list_2):
    return covar(list_1, list_2)/(stdev(list_1)*stdev(list_2))


print(dev([1, 5, 17, 26, 27, 35, 38]))
print(var([1, 5, 17, 26, 27, 35, 38]))
print(stdev([1, 5, 17, 26, 27, 35, 38]))
```
