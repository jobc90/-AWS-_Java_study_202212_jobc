## 전국 로컬푸드 인증정보 표준데이터

- ndarray 사용가능 (문자열 데이터 타입)

- 품목명 top 10(품목별 카운트)
    - 다중품목(+ 또는, 로 연결)은 각각 분리
    - 이외 단일품목으로 취급.
    - 나머지는 알아서. ex)파(실파+대파) --> 파 또는 [실파,대파]
- 주소 공란 : 옆칸 조회해서 메꾸기
- dictionary 작성: {군단위 지역명: [품목명]}
- dictionary 작성: {품목 : 군단위 지역명}
    
- jupyter note 업로드: 이름_로컬푸드.ipymb

```python
import numpy as np

csv_data = []
with open ('전국로컬푸드인증정보표준데이터.csv') as f:
    for line in f:
        t = line[:-1].split(',')
        csv_data.append(t)   
print(csv_data[0])
print(csv_data[1])
csv_data = csv_data[1:]
```

```python
print(csv_data[19][4])
print(csv_data[19][6])
print(csv_data[19][7])
```

```python
adress_list = []

for i in csv_data:
    if(i[4] != ''):
        adress_list.append(i[4].replace(',', '').replace('"', ''))
    elif(i[6] != ''):
        adress_list.append(i[6].replace(',', '').replace('"', ''))
    elif(i[7] != ''):
        adress_list.append(i[7].replace(',', '').replace('"', ''))
    else:
        pass

print(len(adress_list))
print(adress_list)
```

```python
ad_list = []
city_list = []
for i in adress_list:
    ad_list.append(i.split(' '))

for i in ad_list:
    city_list.append(i[0] +' '+i[1])
#print(len(city_list))
#print(city_list)
# for n in set(city_list):
#     print(n, city_list.count(n))


for i in range(len(city_list)):
    if city_list[i].startswith("영월군"):
        city_list[i] = '강원도 영월군'
    elif city_list[i].startswith("강원도 영월읍"):
        city_list[i] = '강원도 영월군'
    elif city_list[i].startswith("강원도 북면"):
        city_list[i] = '강원도 영월군'
    elif city_list[i].startswith("충청북도 옥천군안내면"):
        city_list[i] = '충청북도 옥천군'
all_city_list = []
all_city_list = city_list


city_name = []
city_count = []

for n in set(all_city_list):
    #print(n, city_list.count(n))
    city_name.append(n)
    city_count.append(city_list.count(n))



#print(city_name)
#print(city_count)



city_dict = dict(zip(city_name, city_count))
city_dict = dict(sorted(city_dict.items(), key=lambda x: x[1], reverse=True))
print(city_dict)

# x = list(city_dict.keys())
```

```python
product_list = []

for i in csv_data:
    
    try:
        product_list.append(i[8])
    except:
        pass
print(len(product_list))
print(product_list)
```

```python
pd_list = []

for i in product_list:
    pd_list.append(i.split('+'))
       
    
print(pd_list)
```

```python
pd_flat = []
for i in pd_list:
    pd_flat = pd_flat + i
    
pd_flat.remove('')

# result = [x for x in pd_flat if "순순순" in x]

# print(result)

pd_filter = []
for i in pd_flat:
    if '참외' in i:
        pd_filter.append('참외')
    else:
        pd_filter.append(i)

pd_filter = [x.split('외 ')[0] for x in pd_filter]    
pd_filter = [x.split('(')[0] for x in pd_filter]
pd_filter = [x.split(')')[0] for x in pd_filter]
pd_filter = [x.split('_')[0] for x in pd_filter]
pd_filter = [x.split('-')[0] for x in pd_filter]


# result2 = [x for x in pd_filter if "외" in x]
# print(result2)


```

```python
pd_name = list(set(pd_filter))
pd_count = []
for num in pd_name:
    x = pd_filter.count(num)
    pd_count.append(x)


pd_dict = dict(zip(pd_name, pd_count))
pd_dict = dict(sorted(pd_dict.items(), key=lambda x: x[1], reverse=True))
print(pd_dict)
```

```python
# 최다 지역, 품목
#print(city_dict)
city_top10 = sorted(city_dict, key=city_dict.get, reverse=True)[:10]
print(city_top10)
for i in city_top10:
    print(i ,":", city_dict[i])
#print(pd_dict)
product_top10 = sorted(pd_dict, key=pd_dict.get, reverse=True)[:10]
print(product_top10)
for i in product_top10:
    print(i ,":", pd_dict[i])
```
