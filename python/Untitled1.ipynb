## 전국 로컬푸드 인증정보 표준데이터

- ndarray 사용가능 (문자열 데이터 타입)

- 품목명 top 10(품목별 카운트)
    - 다중품목(+ 또는, 로 연결)은 각각 분리
    - 이외 단일품목으로 취급.
    - 나머지는 알아서. ex)파(실파+대파) --> 파 또는 [실파,대파]
- 주소 공란 : 옆칸 조회해서 메꾸기
- dictionary 작성: {군단위 지역명: [품목명]}
- dictionary 작성: {품목 : 군단위 지역명}
    
- jupyter note 업로드: 이름_로컬푸드.ipymb

```python
import numpy as np

csv_data = []
with open ('전국로컬푸드인증정보표준데이터.csv') as f:
    for line in f:
        t = line[:-1].split(',')
        csv_data.append(t)   
print(csv_data[0])
print(csv_data[1])
csv_data = csv_data[1:]
```

```python
print(csv_data[55][4])
print(csv_data[55][6])
print(csv_data[55][7])
print(csv_data[55][8])
```

```python
adress_list = []

for i in csv_data:
    if(i[6] != ''):
        adress_list.append(i[6])
    elif(i[7] != ''):
        adress_list.append(i[7])
    elif(i[4] != ''):
        adress_list.append(i[4])
    else:
        pass

print(len(adress_list))
print(adress_list)
```

```python
ad_list = []
city_list = []
for i in adress_list:
    ad_list.append(i.split(' '))

for i in ad_list:
    city_list.append(i[0] +' '+i[1])
print(len(city_list))
print(city_list)
```

```python
product_list = []

for i in csv_data:
    
    try:
        product_list.append(i[8])
    except:
        pass
print(len(product_list))
print(product_list)
```
