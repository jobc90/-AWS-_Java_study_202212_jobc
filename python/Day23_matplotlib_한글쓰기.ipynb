```python
import numpy as np
import pandas as pd
import requests
import matplotlib.pyplot as plt
import matplotlib
```

```python
xs = np.linspace(0, 50, 100)
ys = np.sin(xs)

# print(xs)
# print(ys)

figure, axe = plt.subplots()
figure.set_facecolor('red')
figure.suptitle('my axe title')
figure.supxlabel('Figure_x')
figure.supylabel('Figure_y')
# figure.xlim(0, 30)
# figure.ylim(0, 1)
figure.xticks(np.linspace(0, 30, 5))
# figure.xticklables(list('abcde'))

axe.plot(xs, ys)
print(type(figure))
print(type(axe))
```
