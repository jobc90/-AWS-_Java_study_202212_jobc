### 데이터산업의 기술등급별 데이터직무 인력 현황

* 데이터 직무별 axe 생성
    - x축: 시점
    - y축: 업종(2개)
    
* 하나의 figure에, 총 8개 axe가 바둑판 형식으로 배치되는 형태. (4 col x 2 rows), figure 사이즈는 각자 알아서
* title 표시

```python
import numpy as np
import pandas as pd
import requests
import matplotlib
import matplotlib.pyplot as plt

font_path = "C:\Windows\Fonts\HMFMMUEX.TTC"
kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()
#print(kfont)

matplotlib.rcParams['font.family'] = [kfont]


pd_data = pd.read_csv('데이터산업의_기술등급별_데이터직무_인력_현황_20230303160415.csv', encoding='cp949', header=0)



#pd_data.columns = ['데이터직무별(1)', '시점', '데이터 처리 및 관리 솔루션 개발·공급업', '데이터 구축 및 컨설팅 서비스업']
pd_data
```

```python
filter1 = pd_data['데이터직무별(1)']=='데이터 아키텍트'
sub_data = pd_data[filter1]

_, axe = plt.subplots()
axe.plot(sub_data['시점'], sub_data.iloc[:,3])
axe.plot(sub_data['시점'], sub_data.iloc[:,4])
```

```python
sub_data = pd_data.groupby('데이터직무별(1)')
```

```python
sub_data1 = sub_data.get_group('데이터 아키텍트')
_, axe = plt.subplots()
axe.plot(sub_data1['시점'], sub_data1.iloc[:,3])
axe.plot(sub_data1['시점'], sub_data1.iloc[:,4])
```

```python
sub_data2 = sub_data.get_group('데이터 개발자')
_, axe = plt.subplots()
axe.plot(sub_data2['시점'], sub_data2.iloc[:,3])
axe.plot(sub_data2['시점'], sub_data2.iloc[:,4])
```

```python
sub_data3 = sub_data.get_group('데이터 엔지니어')
_, axe = plt.subplots()
axe.plot(sub_data3['시점'], sub_data3.iloc[:,3])
axe.plot(sub_data3['시점'], sub_data3.iloc[:,4])
```

```python
sub_data4 = sub_data.get_group('데이터 분석가')
_, axe = plt.subplots()
axe.plot(sub_data4['시점'], sub_data4.iloc[:,3])
axe.plot(sub_data4['시점'], sub_data4.iloc[:,4])
```

```python
sub_data5 = sub_data.get_group('데이터 분석가')
_, axe = plt.subplots()
axe.plot(sub_data5['시점'], sub_data5.iloc[:,3])
axe.plot(sub_data5['시점'], sub_data5.iloc[:,4])
```

```python
lista = list(sub_data.groups.keys())
print(lista)
```

```python
for n in lista:
    my_data = sub_data.get_group(n)
    fig, axe = plt.subplots()
    axe.plot(my_data['시점'], my_data.iloc[:,3])
    axe.plot(my_data['시점'], my_data.iloc[:,4])
    axe.set_title(n)  # set the title to the group name
    axe.set_xlabel('년도')
    axe.set_ylabel('업체 수')
    axe.set_xticks(np.linspace(2018, 2021, 4))
```

```python
figa = plt.figure(figsize=(17,22))
a = figa.add_subplot(4,2,1)
b = figa.add_subplot(4,2,2)
c = figa.add_subplot(4,2,3)
d = figa.add_subplot(4,2,4)
e = figa.add_subplot(4,2,5)
f = figa.add_subplot(4,2,6)
g = figa.add_subplot(4,2,7)
h = figa.add_subplot(4,2,8)

for n, axe in zip(lista, [a,b,c,d,e,f,g,h]):
    my_data = sub_data.get_group(n)
    axe.plot(my_data['시점'], my_data.iloc[:,3], label='개발·공급업')
    axe.plot(my_data['시점'], my_data.iloc[:,4], label='데이터 구축 및 컨설팅')
    axe.set_title(n)  # set the title to the group name
    axe.set_xlabel('년도')
    axe.set_ylabel('인력 현황')
    axe.set_xticks(np.linspace(2018, 2021, 4))
    axe.legend()
    axe.set_ylim(0, 25000)
```
