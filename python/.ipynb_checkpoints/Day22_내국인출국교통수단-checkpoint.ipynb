```python
import numpy as np
import pandas as pd
import requests
import matplotlib.pyplot as plt

pd_data = pd.read_csv('내국인출국교통수단별_20230302165102.csv', encoding='cp949', header=1)



pd_data.drop('계', axis =1, inplace=True)
pd_data.head()

pd_data.columns = ['시점', '공항', '항구']
pd_data.head()

# 공항, 항구 년중 월별 출국 인원 plot
# normalization / standardization 적용, plot
# 2018부터 2022년까지 추이 보기.
```

```python
_, axe = plt.subplots()
axe.plot(pd_data['공항'])
axe.plot(pd_data['항구'])
```

```python
pd_data['공항'].describe()
bmi_min = pd_data['공항'].min()
bmi_max = pd_data['공항'].max()
pd_data['공항_norm'] = pd_data['공항'].apply(lambda x: (x-bmi_min)/(bmi_max-bmi_min))

pd_data['항구'].describe()
bmi_min = pd_data['항구'].min()
bmi_max = pd_data['항구'].max()
pd_data['항구_norm'] = pd_data['항구'].apply(lambda x: (x-bmi_min)/(bmi_max-bmi_min))

_, axe = plt.subplots()
axe.plot(pd_data['공항_norm'])
axe.plot(pd_data['항구_norm'])
```

```python
pd_data['공항'].describe()
bmi_mean = pd_data['공항'].mean()
bmi_std = pd_data['공항'].std()
pd_data['공항_std'] = pd_data['공항'].apply(lambda x: (x-bmi_mean)/bmi_std)
#print(pd_data['공항_std'])

pd_data['항구'].describe()
bmi_mean = pd_data['항구'].mean()
bmi_std = pd_data['항구'].std()
pd_data['항구_std'] = pd_data['항구'].apply(lambda x: (x-bmi_mean)/bmi_std)
#print(pd_data['항구_std'])

_, axe = plt.subplots()
axe.plot(pd_data['공항_std'])
axe.plot(pd_data['항구_std'])
```

```python
df = pd.DataFrame(df_combined)

# Extract year from the '시점' column and group by year
df['Year'] = pd.DatetimeIndex(df['시점']).year
grouped = df.groupby('Year')

# Aggregate the sum of the columns for each year
result = grouped.agg({'공항': 'sum', '항구': 'sum'})

# Rename the index and show the result
result.index.names = ['시점']
print(result)
```
