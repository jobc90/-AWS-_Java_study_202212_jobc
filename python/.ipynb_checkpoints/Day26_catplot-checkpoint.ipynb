```python
import numpy as np
import pandas as pd
import requests
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

pd_tips = sns.load_dataset('tips')
print(pd_tips)
```

```python
sns.catplot(data=pd_tips, x="day", y="total_bill")

sns.relplot(data=pd_tips, x="day", y="total_bill")
```

```python
sns.catplot(data=pd_tips.query("size != 3"), x="day", y="total_bill", jitter=0.2, hue='smoker', dodge=True)
```

```python
pd_tips["weekend"] = pd_tips["day"].isin(["Sat", "Sun"])
sns.catplot(
    data=pd_tips, x="day", y="total_bill", hue="weekend",
    kind="box", dodge=False,
)
```
