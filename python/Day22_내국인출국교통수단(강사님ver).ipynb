```python
import numpy as np
import pandas as pd
import requests
import matplotlib.pyplot as plt

pd_data = pd.read_csv('내국인출국교통수단별_20230302165102.csv', encoding='cp949', header=1)



pd_data.drop('계', axis =1, inplace=True)
pd_data.head()

pd_data.columns = ['시점', '공항', '항구']
pd_data.head()

# 공항, 항구 년중 월별 출국 인원 plot
# normalization / standardization 적용, plot
# 2018부터 2022년까지 추이 보기.
```

```python
sub_data2022 = pd_data.tail(12)
print(sub_data2022)
print(sub_data2022.describe())

value_min = sub_data2022['공항'].min()
value_max = sub_data2022['공항'].max()
value_std = sub_data2022['공항'].std()
value_mean = sub_data2022['공항'].mean()
sub_data2022['공항_norm'] = sub_data2022['공항'].apply(lambda x: (x-value_min)/(value_max-value_min))
sub_data2022['공항_std'] = sub_data2022['공항'].apply(lambda x: (x-value_mean)/value_std)

value_min = sub_data2022['항구'].min()
value_max = sub_data2022['항구'].max()
value_std = sub_data2022['항구'].std()
value_mean = sub_data2022['항구'].mean()
sub_data2022['항구_norm'] = sub_data2022['항구'].apply(lambda x: (x-value_min)/(value_max-value_min))
sub_data2022['항구_std'] = sub_data2022['항구'].apply(lambda x: (x-value_mean)/value_std)




_, axe = plt.subplots()
axe.plot(np.arange(1, 13), sub_data2022['공항_norm'])
axe.plot(np.arange(1, 13), sub_data2022['항구_norm'])

_, axe = plt.subplots()
axe.plot(np.arange(1, 13), sub_data2022['공항_std'])
axe.plot(np.arange(1, 13), sub_data2022['항구_std'])
```

```python
df_combined = pd.read_csv('내국인출국교통수단별_2.csv', encoding='cp949', header=1)



df_combined.drop('계', axis =1, inplace=True)
df_combined.head()

df_combined.columns = ['시점', '공항', '항구']
print(df_combined)


# 공항, 항구 년중 월별 출국 인원 plot
# normalization / standardization 적용, plot
# 2018부터 2022년까지 추이 보기.
```

```python
print(df_combined)
for i in range(8):
    #print(i, 12*i, 12*(i+1))
    sub_data = df_combined.iloc[12*i:12*(1+i)]
    #print(sub_data)
    
    for n in ['공항', '항구']:
        value_min = sub_data[n].min()
        value_max = sub_data[n].max()
        value_std = sub_data[n].std()
        value_mean = sub_data[n].mean()
        sub_data[n+'_norm'] = sub_data[n].apply(lambda x: (x-value_min)/(value_max-value_min))
        sub_data[n+'_std'] = sub_data[n].apply(lambda x: (x-value_mean)/value_std)
    
    _, axe = plt.subplots()
    axe.plot(np.arange(1, 13), sub_data['공항'])
    axe.plot(np.arange(1, 13), sub_data['항구'])
    
    _, axe = plt.subplots()
    axe.plot(np.arange(1, 13), sub_data['공항_norm'])
    axe.plot(np.arange(1, 13), sub_data['항구_norm'])

    _, axe = plt.subplots()
    axe.plot(np.arange(1, 13), sub_data['공항_std'])
    axe.plot(np.arange(1, 13), sub_data['항구_std'])
```
